```python
# Create the initial, empty board with three rows and three columns.
# Create a function to reset the board.

row1 = [" ", " ", " "]
row2 = [" ", " ", " "]
row3 = [" ", " ", " "]
my_board = [row1, row2, row3]

def print_board():
    print(my_board[0])
    print(my_board[1])
    print(my_board[2])

def reset_board():
    row1 = [" ", " ", " "]
    row2 = [" ", " ", " "]
    row3 = [" ", " ", " "]
    my_board = [row1, row2, row3]
    
    return my_board
```

```python
# Input functions: grabs name and symbol for each player and stores it in two lists:
# [names] and [symbols]. 

def capture_names_symbols():
    
    names = []
    symbols = []
    
    symbol = ""
    
    for i in range(2):
        name = (input("Please enter your name: "))
        names.append(name)
            
        while True:
            symbol = input("Please enter the symbol you want to use (O or X): ")
        
            if symbol not in ["O", "X"]:
                print("Sorry, I don't recognize that symbol. Please try again, choose O or X")
            
            else:
                symbols.append(symbol)
                
                if i > 0 and symbol == symbols[i-1]:
                    symbols.pop()
                    print("Sorry, {}, that symbol is taken. Please, pick the other one.".format(names[i]))
                    continue
                break
    return names, symbols
```

```python
# Function that checks whether the cell chosen is empty.

def is_empty():
    
    while my_board[row][col] != " ":
        if my_board[row][col] != " ":
            print("Sorry {}, the cell is TAKEN. Please pick another one.".format(current_player))
            user_inputs()
        else:
            print("Sounds about right!")
        
```

```python
# Function that updates the board, prints it, and updates the current player.

def update_board_player():
    
    global current_player
    
    if current_player == names[0]:
        my_board[row][col] = symbols[0]
        current_player = names[1]
    else:
        my_board[row][col] = symbols[1]
        current_player = names[0]
        
    print_board()
    return current_player
```

```python
# Function that checks if there's a winner.
# Checks for three in a row, column or diagonal. Returns the boolean 'checker' as T or F.

def is_there_winner():
    global winner_symbol
    checker = False
    winner_symbol = ""

    # Check DIAGONAL wins

    i = 0
    
    if (my_board[i][i]==my_board[i+1][i+1] == my_board[i+2][i+2]) and my_board[i][i] in symbols:
        print("WE HAVE A WINNER in diagonal!")
        
        checker = True
        winner_symbol = my_board[i][i]
              
    elif (my_board[i][i+2]==my_board[i+1][i+1] == my_board[i+2][i]) and my_board[i][i+2] in symbols:
        print("WE HAVE A WINNER in diagonal!")
        checker = True
        winner_symbol = my_board[i][i+2]
              
     # Check ROW and COLUMN wins
    for n in range(3):
        i = 0

        if (my_board[n][i]==my_board[n][i+1] == my_board[n][i+2]) and my_board[n][i] in symbols:
            print("WE HAVE A WINNER in a row!")
            checker = True
            winner_symbol = my_board[n][i]

        elif (my_board[i][n]==my_board[i+1][n] == my_board[i+2][n]) and my_board[i][n] in symbols:
            print("WE HAVE A WINNER in a column!")
            checker = True
            winner_symbol = my_board[i][n]
                  
    return checker
```

```python
# Function that reveals the winner's name when there is one.

def reveal_winner():

  if winner_symbol == symbols[0]:
    print("The winner is {}".format(names[0]))
  else:
    print("The winner is {}".format(names[1]))
```

```python
# Function that checks if the board is full.

def board_full():
    counter = 0
    
    for eachrow in my_board:
        for eachcell in eachrow:
            if eachcell == " ":
                counter += 1
    if counter == 0:
        return True
    else:
        print(counter)
        return False
```
