## This short Python project can be used to randomly draw names from a list. It accepts a CSV file with names and number of entries, generates the list based on those parameters, and performs a random draw.

```python
import csv
import random

# Function that generates the full list by multiplying the each name for the number
# of entries (raffle tickets) purchased by that person.

def full_list_generator(spreadsheet):
    with open(spreadsheet, 'r') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')

        names = []
        full_list = []

        for row in csv_reader:
            names.append(row[0])
            multiply = (row[0] + "-") * int(row[1])
            full_list.append(multiply.strip().split("-"))
    
    return full_list

# Function that merges everything into one list and filters out empty entries.

def clean_list(full_list):
    
    concat_list = [j for i in full_list for j in i]

    filter_object = filter(lambda x: x != "", concat_list)
    final_list = list(filter_object)
    
    return final_list

# Function that performs the random draw: selects a name based on a randomly
# generated integer between 0 and len(list), displays it, and then removes it
# from the list.

def draw(final_list):
    
    card_index = random.randrange(0, len(final_list))
    print("THE WINNER IS: ", final_list[card_index])
    final_list.pop(card_index)
    print("Remaining entries: ", len(final_list))
    
    return final_list
```

```python
# Generating and cleaning up the list based on our CSV file

full_list = full_list_generator('random-draw-test.csv')

final_list = clean_list(full_list)
```

```python
draw(final_list)
```
